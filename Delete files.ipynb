```python
import glob

firstDay = ['mayroon', 'kayo', 'black', 'ano', 'niyo', 'kape', 'magkano', 'bibili', 'oras', 'tubig', 'maliit (thing)', 
            'kulay', 'isa', 'dito', 'pagkain']
secondDay1 = ['nagbubukas', 'ate', 'isda', 'itlog', 'kuya', 'baboy']
secondDay2 = ['t-shirt', 'blue', 'wala', 'dalawa', 'tatlo', 'nagsasara', 'shorts', 'red', 'mahal', 
              'tawad', 'small (shirt)', 'medium (shirt)', 'large (shirt)', 'sari-sari store', 'mantika', 'juice', 'size']
thirdDay = ['saging', 'nito', 'kanin', 'ito', 'spaghetti', 'gusto', 'tinapay', 'manok', 'poloshirt', 'extra', 'kilo', 'salamat', 'binebenta', 'puwede', 'welcome', 'utang', 'gulay',
           'softdrinks', 'sorry', 'hello', 'malaki', 'supot (plastik)']

listOfFilesPerClass = {}

for c in listOfClassNames:
    files = glob.glob("/Users/user/Desktop/Thesis/Data/placeholder/"+c+"/*.csv")
    listOfFilesPerClass[c] = {}
    print(len(files))
    listOfRightEMG = []
    listOfLeftEMG = []
    listOfRightACC = []
    listOfLeftACC = []
    listOfRightGYRO = []
    listOfLeftGYRO = []
    
    for fileName in files:
        if "emg" in fileName:   
            if "-0-" in fileName:
                listOfRightEMG.append(fileName)
            else:
                listOfLeftEMG.append(fileName)
        elif "accelerometer" in fileName:   
            if "-0-" in fileName:
                listOfRightACC.append(fileName)
            else:
                listOfLeftACC.append(fileName)
        elif "gyro" in fileName:   
            if "-0-" in fileName:
                listOfRightGYRO.append(fileName)
            else:
                listOfLeftGYRO.append(fileName)
    
    listOfFilesPerClass[c]['remg'] = listOfRightEMG
    listOfFilesPerClass[c]['lemg'] = listOfLeftEMG
    listOfFilesPerClass[c]['racc'] = listOfRightACC
    listOfFilesPerClass[c]['lacc'] = listOfLeftACC
    listOfFilesPerClass[c]['rgyro'] = listOfRightGYRO
    listOfFilesPerClass[c]['lgyro'] = listOfLeftGYRO
```
